2.3.1 Getting Shape and Size of an Image Using PIL

For Gray Scale Image:

```python
# Get the size (width , height ) and channels (RGB)
width , height = image_grayed.size
channels = len( image_grayed . getbands () ) # For RGB , it will be 3
print ( f" Image shape (RGB): ({ height } , { width } , { channels})")
image_size_grayed = width * height *1
print ( f" Image size (RGB): { image_size_grayed }")
```

2.5.1 Extracting Colored Channels with NumPy Array.

Accessing individual channels

```python
# Convert the Pillow image to a NumPy array
image_array_colored = np.array ( image_colored )
# Display the shape of the NumPy array (height , width ,channels )
print (" Shape of the image array :", image_array_colored.shape)
#For Red Channels
red_channel = image_array_colored [243, 0] # Red channel
print(red_channel)
display(red_channel)
# For Green Channels
green_channel = image_array_colored [345, 0] # Green channel
print(green_channel)
display(green_channel)
```
